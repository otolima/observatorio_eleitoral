```python
import pandas as pd
from sqlalchemy import create_engine, text
import requests
import zipfile
import io

# 1. Conexão
engine = create_engine('postgresql://admin:senha_eleicoes_2026@localhost:5432/db_eleicoes')

# 2. URL Oficial do TSE (Dados de Candidatos 2022 - Brasil)
# Esta URL é mais estável por ser um repositório histórico
url = "https://cdn.tse.jus.br/estatistica/sead/odsele/consulta_cand/consulta_cand_2022.zip"

try:
    print("⏳ Passo 1: Baixando arquivo ZIP do TSE... (Isso pode levar 1 minuto)")
    r = requests.get(url, verify=False) # verify=False evita erros de certificado SSL comuns em órgãos públicos
    
    # Lendo o conteúdo do ZIP
    z = zipfile.ZipFile(io.BytesIO(r.content))
    
    # Listar arquivos dentro do ZIP e pegar o CSV do Brasil
    # O arquivo principal geralmente termina com _BRASIL.csv
    csv_file = [f for f in z.namelist() if f.endswith('_BRASIL.csv')][0]
    
    print(f"📂 Extraindo arquivo: {csv_file}")
    
    # Lendo o CSV (especificando o encoding 'latin-1' que o TSE usa)
    df_raw = pd.read_csv(z.open(csv_file), sep=';', encoding='latin-1')
    
    print(f"✅ Dados lidos com sucesso! Total de linhas: {len(df_raw)}")

    # 3. Criar Schema e Carregar no PostgreSQL
    with engine.connect() as conn:
        conn.execute(text("CREATE SCHEMA IF NOT EXISTS bronze;"))
        conn.commit()

    print("📤 Enviando para a camada Bronze no Docker...")
    df_raw.to_sql('candidatos_raw', engine, schema='bronze', if_exists='replace', index=False)
    
    print("🚀 SUCESSO! Dados carregados na tabela bronze.candidatos_raw")
    display(df_raw.head())

except Exception as e:
    print(f"❌ Erro: {e}")
```
