```python
import pandas as pd
from sqlalchemy import create_engine
import requests
import zipfile
import io

# 1. Conexão
engine = create_engine('postgresql://admin:senha_eleicoes_2026@localhost:5432/db_eleicoes')

# 2. URL
url_bens = "https://cdn.tse.jus.br/estatistica/sead/odsele/bem_candidato/bem_candidato_2024.zip"

print("⏳ Baixando e extraindo o arquivo BRASIL... Aguarde.")

# 3. Baixando o conteúdo para a memória
r = requests.get(url_bens)
z = zipfile.ZipFile(io.BytesIO(r.content))

# 4. Lendo especificamente o arquivo consolidado do Brasil
# O nome do arquivo dentro do ZIP é 'bem_candidato_2024_BRASIL.csv'
with z.open('bem_candidato_2024_BRASIL.csv') as f:
    df_bens_raw = pd.read_csv(f, sep=';', encoding='latin-1')

print(f"✅ Arquivo BRASIL carregado! Registros: {len(df_bens_raw)}")

# 5. Salvando na Bronze
df_bens_raw.to_sql('bens_raw', engine, schema='bronze', if_exists='replace', index=False)

print("🚀 Tabela 'bronze.bens_raw' criada com sucesso!")
```

```python
# 1. Padronizar colunas para minúsculo
df_bens_raw.columns = [col.lower() for col in df_bens_raw.columns]

# 2. Selecionar apenas o essencial para a Silver
# sq_candidato é o nosso "grampo" para o JOIN futuro
colunas_bens = ['sq_candidato', 'ds_tipo_bem_candidato', 'detalhe_bem', 'vr_bem']
df_bens_silver = df_bens_raw[colunas_bens].copy()

# 3. Mágica da Limpeza Financeira:
# Troca a vírgula por ponto e transforma em número (float)
df_bens_silver['vr_bem'] = df_bens_silver['vr_bem'].str.replace(',', '.').astype(float)

# 4. Criar o Schema Silver (caso ainda não exista para esta sessão)
from sqlalchemy import text
with engine.connect() as conn:
    conn.execute(text("CREATE SCHEMA IF NOT EXISTS silver;"))
    conn.commit()

# 5. Salvar na Tabela Silver
df_bens_silver.to_sql('bens_silver', engine, schema='silver', if_exists='replace', index=False)

print("✨ Camada Silver de Bens concluída com sucesso!")
print(f"Exemplo de valor processado: {df_bens_silver['vr_bem'].iloc[0]}")
```

```python
# 1. Definimos a lista com os nomes EXATOS que você postou acima
colunas_reais = [
    'sq_candidato', 
    'ds_tipo_bem_candidato', 
    'ds_bem_candidato',   # Este é o antigo 'detalhe_bem'
    'vr_bem_candidato'    # Este é o antigo 'vr_bem'
]

# 2. Fazemos a seleção e a cópia (usando o .copy() para segurança)
df_bens_silver = df_bens_raw[colunas_reais].copy()

# 3. Renomeamos para nomes mais curtos e amigáveis para o nosso banco de dados
df_bens_silver.columns = ['sq_candidato', 'tipo_bem', 'detalhe_bem', 'valor_bem']

# 4. Limpeza do Valor: Transformando texto "150.000,00" em número 150000.00
# Primeiro garantimos que é texto, trocamos a vírgula por ponto e convertemos para float
df_bens_silver['valor_bem'] = df_bens_silver['valor_bem'].astype(str).str.replace(',', '.').astype(float)

# 5. Salvar na Tabela Silver (agora com as colunas certinhas!)
df_bens_silver.to_sql('bens_silver', engine, schema='silver', if_exists='replace', index=False)

print("✨ Sucesso! Camada Silver de Bens criada com os nomes corrigidos.")
```
